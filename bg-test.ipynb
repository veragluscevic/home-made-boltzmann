```python
import numpy as np
import matplotlib.pyplot as plt
import scipy as sp
import astropy as ap
```

```python
def background(a=0, Omega_dm_0=0, Omega_b_0=0, Omega_de_0=0, Omega_r_0=0):
    a = np.atleast_1d(a)
    Omega_dm = Omega_dm_0 / a**3
    Omega_b = Omega_b_0 / a**3
    Omega_r = Omega_r_0 / a**4
    Omega_de = Omega_de_0 *a/a

    Hs = compute_Hubble(a=a, Omega_dm_0=Omega_dm_0, Omega_b_0=Omega_b_0, Omega_de_0=Omega_de_0, Omega_r_0=Omega_r_0)
    bg = {}
    bg = {
        'Omega_dm': Omega_dm,
        'Omega_b': Omega_b,
        'Omega_r': Omega_r,
        'Omega_de': Omega_de,
        'a': a, 
        'H': Hs 
    }

    return bg
    
def compute_Hubble(a=0, H0=70, Omega_dm_0=0, Omega_b_0=0, Omega_de_0=0, Omega_r_0=0):
    """H0 [km/sec/Mpc]"""
    a = np.atleast_1d(a)
    H = H0 * np.sqrt(Omega_dm_0 / a**3 + Omega_b_0 / a**3 + Omega_r_0 / a**4 + Omega_de_0)
    return H
```

```python
a = np.logspace(np.log10(0.5),0,100)
#print(a)
b = {}
b = background (a, 0.25, 0.05, 0.7, 1e-4)
```

```python
plt.loglog(a, b['H'])
```

```python
plt.plot(a, b['Omega_b'])
plt.plot(a, b['Omega_dm'])
plt.plot(a, b['Omega_r'])
```
